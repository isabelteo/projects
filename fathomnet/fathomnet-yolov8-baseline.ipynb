After reformatting to the YoloV8 directory structure, we need to generate the label txts from the COCO json files

```python
import json
import os
import numpy as np

os.chdir(json_path)

with open("train.json", 'r') as train_file:
    train_json = json.load(train_file)
```

```python
# Creates a dictionary where the filename is the key and the values is a dictionary containing information about the image, including annotations
def get_image_filename_dict(coco_json):
    # Reformat for easy access to image data information by id
    image_data = coco_json["images"]
    num_images = len(image_data)
    image_dict = {}
    for image in image_data:
        image_dict[image["id"]] = {"id":image["id"], "width":image['width'], 'height':image['height'],'file_name':image["file_name"], 'annotations':[]}
    
    assert len(image_dict) == num_images
    
    #Add annotation information as list to image_dict from annotations section in COCO json
    annotation_data = coco_json["annotations"]
    for annotation in annotation_data:
        category = annotation["category_id"]
        bbox = annotation["bbox"]
        x, y, w, h = bbox[0], bbox[1], bbox[2], bbox[3]
        
        image_id = annotation["image_id"]
        image_width = image_dict[image_id]["width"]
        image_height = image_dict[image_id]["height"]
        
        x_center = x + w/2
        y_center = y + h/2
        x_center /= image_width
        y_center /= image_height
        w /= image_width
        h /= image_height
            
        yolo_data = [category, x_center, y_center, w, h]
        image_yolo_data = image_dict[image_id]["annotations"] 
        image_yolo_data.append(yolo_data)
        image_dict[image_id]["annotations"] = image_yolo_data
    
    image_filename_dict = {}
    for image_id, image_data in image_dict.items():
        image_filename_dict[image_data["file_name"]] = image_data
    
    return image_filename_dict

#Creates a label txt file for each image in the image_path and outputs the label txt file into the label_path
def create_label_txts(image_filename_dict, image_path, label_path): 
    print(f"Starting generation for images in {image_path} to output in {label_path}")
    for filename in os.listdir(image_path):
        image_data = image_filename_dict[filename]
        assert image_data["annotations"] != []
        image_yolo_data = np.array(image_data["annotations"])
        #remove .png and replace with .txt
        save_as_filename = image_data["file_name"][:-4] + ".txt"
        np.savetxt(label_path + "\\" + save_as_filename, image_yolo_data, fmt=["%d", "%f", "%f", "%f", "%f"])
    print("Done")
    
```

```python
image_filename_dict = get_image_filename_dict(train_json)
```

```python
categories = train_json["categories"]
class_dict = {}
for category in categories:
    #needs to be zero-indexed
    class_dict[category["id"]-1] = category["name"]
    
print(class_dict[0])
```

Start finetuning process

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.pt")  # load a pretrained model

# Training.
results = model.train(
   data='custom.yaml',
   imgsz=640,
   epochs=10,
   batch=16,
   name='yolov8_fathomnet')
```

```python
with open(labels_path, 'r') as labels:
    labels = json.load(labels)
```

```python
for i in range(len(labels["annotations"])):
    cat_id = int(labels["annotations"][i]["category_id"])
    labels["annotations"][i]["category_id"] = cat_id

with open(labels_path, "w") as outfile:
    json.dump(labels, outfile,indent=4)    
```

Upload data (formatted into COCO dataset format) into a dataset object for fiftyone


```python
import fiftyone as fo


# Import the dataset
dataset = fo.Dataset.from_dir(
    dataset_type=fo.types.COCODetectionDataset,
    data_path=data_path,
    labels_path=labels_path,
)

# View summary info about the dataset
print(dataset)

# Print the first few samples in the dataset
print(dataset.head())
```

Compute uniqueness and sort by uniqueness

```python
import fiftyone.brain as fob
import fiftyone.zoo as foz
# Compute your own embeddings
model = foz.load_zoo_model("mobilenet-v2-imagenet-torch")
embeddings = dataset.compute_embeddings(model)

# Index by uniqueness using pre-computed embeddings
fob.compute_uniqueness(dataset, embeddings=embeddings)
print(dataset)

# Show least unique images in the App
session.view = dataset.sort_by("uniqueness")
```
